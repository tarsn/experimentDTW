```python
p1_filtered = {
    'a': [],
    'e': [],
    'i': [],
    'o': [],
    'u': []
}
```

```python
__root_folder__ = "../data/p1/"

for folderNames in os.listdir(__root_folder__):
    for fileNames in os.listdir(__root_folder__ + folderNames):
        if fileNames.endswith(".wav"):
            # print("Processing: ", fileNames)
            filter_audio(input=__root_folder__+folderNames+"/"+fileNames, output="../tmpFiltered/"+fileNames)
            features = FeaturesExtraction(path="../tmpFiltered/", audio_file=fileNames, target=folderNames)
            p1_filtered[folderNames].append(features)
```
